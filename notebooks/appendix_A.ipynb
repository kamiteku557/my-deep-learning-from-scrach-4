```python
import numpy as np
from tqdm.notebook import tqdm

import appendix_A
import ch4

%load_ext autoreload
%autoreload 2
```

```python
# .5
env = ch4.GridWorld()
agent = appendix_A.McOffPolicyAgent()

episodes = 10_000
for episode in tqdm(range(episodes)):
    state = env.reset()
    agent.reset()

    while True:
        action = agent.get_action(state)
        next_state, reward, done = env.step(action)

        agent.add(state, action, reward)
        if done:
            agent.update()
            break

        state = next_state

env.render_q(agent.Q)
```
